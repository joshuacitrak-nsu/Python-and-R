```python
import pandas as pd

df = pd.read_csv("/Users/mattrosinski/Documents/Repos/Demos/Python-and-R/data/company_202209.csv", sep='\t', engine='python')
```

```python
# Convert to datetime format (state format explicitly convert from UK/Australian format)
df['Date of Registration'] =  pd.to_datetime(df['Date of Registration'], format='%d/%m/%Y')

# Create month and year columns
df.insert(1, 'month', df['Date of Registration'].dt.month_name())

df['year'] = pd.DatetimeIndex(df['Date of Registration']).year.astype('Int32')
```

## New Company Registrations by Month in Australia

```python
# Create custom ordered month data type
monthorder = ['January', 'February', 'March', 'April', 'May', 'June', 'July', \
              'August', 'September', 'October', 'November', 'December']
ordered_month_type = pd.CategoricalDtype(categories=monthorder, ordered=True)

df['month'] = df['month'].astype(ordered_month_type)

# Filter out companies with not registration date
filt = df['Date of Registration'].notna()
df = df.loc[filt]

(df.pivot_table(index='year', values='ACN', columns='month', aggfunc='size', )
  .iloc[-21:,:]
  .round(-2)
  .div(1000)
  .style
  .background_gradient(axis=1)
  .highlight_max(axis=None, color='Salmon')
  .format('{:,}k'))
```

## New Company Registrations by Month in Australia

```python
# Alternative using crosstab.
(pd.crosstab(index=df['year'], columns=df['month'])
    .iloc[-10:,:]
    .round(-2)
    .div(1000)
    .style
    .background_gradient(axis=1)
    .highlight_max(axis=None, color='Salmon')
    .format('{:,}k'))
```
